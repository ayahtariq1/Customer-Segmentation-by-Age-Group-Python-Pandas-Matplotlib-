```python

import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from dash import Dash, html, dcc
from dash.dependencies import Input, Output, State

# ================== Load data ==================
df = pd.read_csv("shopping_trends_updated.csv")

# ================== App & Styles ==================
app = Dash(__name__)

# Ø£Ù„ÙˆØ§Ù† Ù…ØªÙ†ÙˆØ¹Ø© ÙˆÙ…Ø±ØªØ¨Ø©
PALETTE = [
    "#2F7AE5","#6C5CE7","#00B894","#E17055","#ED4C67","#10AC84",
    "#A3CB38","#1289A7","#D980FA","#FFC312","#C4E538","#EE5253"
]
BG = "#F8FAFC"
CARD_BG = "#FFFFFF"
TEXT = "#0F172A"
GRID = "#E2E8F0"

card_style = {
    "padding": "16px",
    "borderRadius": "12px",
    "background": CARD_BG,
    "boxShadow": "0 2px 10px rgba(0,0,0,0.06)",
    "textAlign": "center",
    "minWidth": "200px",
    "flex": "1",
    "border": f"1px solid {GRID}",
    "color": TEXT
}

section_title_style = {"margin": "0 0 8px 0", "color": TEXT}

def thousands(n):
    try:
        return f"{int(n):,}"
    except:
        return str(n)

# ================== Filter choices (sorted) ==================
def safe_unique(col):
    return sorted(df[col].dropna().astype(str).unique()) if col in df.columns else []

categories = safe_unique("Category")
seasons    = safe_unique("Season")
genders    = safe_unique("Gender")

# ================== Layout ==================
app.layout = html.Div([
    html.H1("Shopping Trends Dashboard", style={"textAlign": "center", "color": TEXT, "marginTop": "10px"}),

    # ======= FILTERS (larger area + Select All) =======
    html.Div([
        # Category
        html.Div([
            html.Label("Category", style=section_title_style),
            dcc.Dropdown(
                id="filter_category",
                options=[{"label": c, "value": c} for c in categories],
                placeholder="Filter by Category",
                multi=True, style={"width": "100%"}
            ),
            dcc.Checklist(id="select_all_category",
                          options=[{"label": " Select All", "value": "ALL"}],
                          value=[], style={"marginTop": "6px"})
        ], style={"flex": "1", "minWidth": "260px"}),

        # Gender
        html.Div([
            html.Label("Gender", style=section_title_style),
            dcc.Dropdown(
                id="filter_gender",
                options=[{"label": g, "value": g} for g in genders],
                placeholder="Filter by Gender",
                multi=True, style={"width": "100%"}
            ),
            dcc.Checklist(id="select_all_gender",
                          options=[{"label": " Select All", "value": "ALL"}],
                          value=[], style={"marginTop": "6px"})
        ], style={"flex": "1", "minWidth": "200px"}),

        # Season
        html.Div([
            html.Label("Season", style=section_title_style),
            dcc.Dropdown(
                id="filter_season",
                options=[{"label": s, "value": s} for s in seasons],
                placeholder="Filter by Season",
                multi=True, style={"width": "100%"}
            ),
            dcc.Checklist(id="select_all_season",
                          options=[{"label": " Select All", "value": "ALL"}],
                          value=[], style={"marginTop": "6px"})
        ], style={"flex": "1", "minWidth": "220px"}),
    ], style={
        "display": "flex", "gap": "20px", "flexWrap": "wrap",
        "background": BG, "padding": "16px", "borderRadius": "12px",
        "border": f"1px solid {GRID}", "margin": "10px 0"
    }),

    html.Br(),

    # ======= KPIs ROW =======
    html.Div(id="kpi_row", style={"display": "flex", "gap": "16px", "flexWrap": "wrap"}),

    html.Br(),

    # ======= TABS with diversified charts =======
    dcc.Tabs([
        dcc.Tab(label="Overview", children=[
            html.Div([
                dcc.Graph(id="sales_by_category"),
                dcc.Graph(id="gender_pie"),
            ], style={"display": "grid", "gridTemplateColumns": "1fr 1fr", "gap": "12px"}),
            html.Div([
                dcc.Graph(id="purchase_season"),           # Polar (Ù…ØªÙ†ÙˆØ¹)
                dcc.Graph(id="payment_method_donut"),      # Donut
            ], style={"display": "grid", "gridTemplateColumns": "1fr 1fr", "gap": "12px"}),
        ]),
        dcc.Tab(label="Product Insights", children=[
            html.Div([
                dcc.Graph(id="rating_by_category"),        # Bar multi-color
                dcc.Graph(id="treemap_category_item"),     # Treemap
            ], style={"display": "grid", "gridTemplateColumns": "1fr 1fr", "gap": "12px"}),
            html.Div([
                dcc.Graph(id="box_amount_by_category"),    # Box plot
                dcc.Graph(id="heatmap_season_category"),   # Heatmap
            ], style={"display": "grid", "gridTemplateColumns": "1fr 1fr", "gap": "12px"}),
        ]),
        dcc.Tab(label="Customer Behavior", children=[
            html.Div([
                dcc.Graph(id="scatter_age_amount"),        # Scatter + marginals
                dcc.Graph(id="violin_amount_by_gender"),   # Violin
            ], style={"display": "grid", "gridTemplateColumns": "1fr 1fr", "gap": "12px"}),
        ]),
    ], style={"marginTop": "4px"})
], style={"padding": "12px", "background": "#FFFFFF", "fontFamily": "Arial, sans-serif"})

# ================== Helpers ==================
def apply_filters(_df, cats, gens, seas):
    dff = _df.copy()
    if "Category" in dff.columns and cats:
        dff = dff[dff["Category"].astype(str).isin(cats)]
    if "Gender" in dff.columns and gens:
        dff = dff[dff["Gender"].astype(str).isin(gens)]
    if "Season" in dff.columns and seas:
        dff = dff[dff["Season"].astype(str).isin(seas)]
    return dff

def style_common(fig, x_title="", y_title=""):
    fig.update_layout(
        template="plotly_white",
        plot_bgcolor="#FFFFFF",
        paper_bgcolor="#FFFFFF",
        hovermode="closest",
        margin=dict(l=40, r=20, t=60, b=60),
        xaxis_title=x_title,
        yaxis_title=y_title,
        font=dict(color=TEXT),
    )
    fig.update_yaxes(gridcolor=GRID, zeroline=False)
    fig.update_xaxes(showgrid=False)
    return fig

def style_pie(fig):
    fig.update_traces(textposition='inside', textinfo='percent+label')
    fig.update_layout(template="plotly_white", margin=dict(l=20, r=20, t=60, b=20))
    return fig

def annotation_fig(msg):
    _fig = px.scatter(x=[0], y=[0])
    _fig.update_traces(marker_opacity=0)
    _fig.update_layout(
        template="plotly_white",
        annotations=[dict(text=msg, x=0.5, y=0.5, showarrow=False, font=dict(size=16, color="#64748B"))],
        xaxis=dict(visible=False), yaxis=dict(visible=False), margin=dict(l=0, r=0, t=40, b=0)
    )
    return _fig

# ================== CALLBACK ==================
@app.callback(
    [
        Output("kpi_row", "children"),
        Output("sales_by_category", "figure"),
        Output("gender_pie", "figure"),
        Output("rating_by_category", "figure"),
        Output("purchase_season", "figure"),
        Output("payment_method_donut", "figure"),
        Output("treemap_category_item", "figure"),
        Output("box_amount_by_category", "figure"),
        Output("heatmap_season_category", "figure"),
        Output("scatter_age_amount", "figure"),
        Output("violin_amount_by_gender", "figure"),
        # keep filters in sync with "Select All"
        Output("filter_category", "value"),
        Output("filter_gender", "value"),
        Output("filter_season", "value"),
    ],
    [
        Input("filter_category", "value"),
        Input("filter_gender", "value"),
        Input("filter_season", "value"),
        Input("select_all_category", "value"),
        Input("select_all_gender", "value"),
        Input("select_all_season", "value"),
    ],
    prevent_initial_call=False
)
def update_dashboard(cat_vals, gen_vals, sea_vals, sel_cat_all, sel_gen_all, sel_sea_all):
    # Handle Select All toggles
    cat_vals = categories if ("ALL" in (sel_cat_all or [])) else (cat_vals or [])
    gen_vals = genders    if ("ALL" in (sel_gen_all or [])) else (gen_vals or [])
    sea_vals = seasons    if ("ALL" in (sel_sea_all or [])) else (sea_vals or [])

    dff = apply_filters(df, cat_vals, gen_vals, sea_vals)

    # ===== KPIs =====
    has_amount = "Purchase Amount (USD)" in dff.columns
    has_rating = "Review Rating" in dff.columns
    has_customer = "Customer ID" in dff.columns

    total_customers = dff["Customer ID"].nunique() if has_customer else len(dff)
    total_revenue = float(dff["Purchase Amount (USD)"].sum()) if has_amount else 0.0
    avg_purchase = float(dff["Purchase Amount (USD)"].mean()) if has_amount else 0.0
    avg_rating = round(float(dff["Review Rating"].mean()), 2) if has_rating else None

    kpi_cards = [
        html.Div([html.H4("Total Customers"), html.H2(thousands(total_customers))], style=card_style),
        html.Div([html.H4("Total Revenue"), html.H2(f"${total_revenue:,.0f}")], style=card_style),
        html.Div([html.H4("Avg Order Value"), html.H2(f"${avg_purchase:,.2f}")], style=card_style),
        html.Div([html.H4("Avg Rating"), html.H2(avg_rating if avg_rating is not None else "N/A")], style=card_style),
    ]

    # ===== Charts =====

    # 1) Sales by Category (Bar multi-color)
    if "Category" in dff.columns and has_amount:
        cat_agg = dff.groupby("Category", as_index=False)["Purchase Amount (USD)"].sum()
        cat_agg = cat_agg.sort_values("Purchase Amount (USD)", ascending=False)
        fig_cat = px.bar(
            cat_agg, x="Category", y="Purchase Amount (USD)",
            color="Category", color_discrete_sequence=PALETTE,
            title="Sales by Category"
        )
        fig_cat.update_traces(hovertemplate="<b>%{x}</b><br>Revenue: $%{y:,.0f}<extra></extra>")
        fig_cat = style_common(fig_cat, x_title="Category", y_title="Revenue (USD)")
    else:
        fig_cat = annotation_fig("Category / Purchase Amount (USD) ØºÙŠØ± Ù…ØªÙˆÙØ±ÙŠÙ†")

    # 2) Gender Donut
    if "Gender" in dff.columns:
        gender_agg = dff["Gender"].value_counts().reset_index()
        gender_agg.columns = ["Gender", "Count"]
        fig_gender = px.pie(
            gender_agg, names="Gender", values="Count", hole=0.5,
            color="Gender", color_discrete_sequence=PALETTE,
            title="Gender Distribution"
        )
        fig_gender = style_pie(fig_gender)
    else:
        fig_gender = annotation_fig("Gender ØºÙŠØ± Ù…ØªÙˆÙØ±")

    # 3) Average Rating by Category (Bar multi-color)
    if "Category" in dff.columns and has_rating:
        rating_agg = dff.groupby("Category", as_index=False)["Review Rating"].mean()
        rating_agg = rating_agg.sort_values("Review Rating", ascending=False)
        fig_rating = px.bar(
            rating_agg, x="Category", y="Review Rating",
            color="Category", color_discrete_sequence=PALETTE,
            title="Average Rating by Category"
        )
        fig_rating.update_traces(hovertemplate="<b>%{x}</b><br>Avg Rating: %{y:.2f}<extra></extra>")
        fig_rating = style_common(fig_rating, x_title="Category", y_title="Avg Rating")
        fig_rating.update_yaxes(range=[0, max(5, rating_agg["Review Rating"].max()+0.2)])
    else:
        fig_rating = annotation_fig("Category / Review Rating ØºÙŠØ± Ù…ØªÙˆÙØ±ÙŠÙ†")

    # 4) Sales by Season (Polar Ù…ØªÙ†ÙˆØ¹)
    if "Season" in dff.columns and has_amount:
        season_agg = dff.groupby("Season", as_index=False)["Purchase Amount (USD)"].sum()
        season_order = ["Winter", "Spring", "Summer", "Fall"]
        season_agg["Season"] = pd.Categorical(season_agg["Season"], categories=season_order, ordered=True)
        season_agg = season_agg.sort_values("Season")
        fig_season = px.bar_polar(
            season_agg, r="Purchase Amount (USD)", theta="Season",
            color="Season", color_discrete_sequence=PALETTE,
            title="Revenue by Season (Polar)"
        )
        fig_season.update_layout(template="plotly_white", margin=dict(l=40, r=20, t=60, b=40))
        fig_season.update_polars(radialaxis=dict(gridcolor=GRID, showticklabels=True))
    else:
        fig_season = annotation_fig("Season / Purchase Amount (USD) ØºÙŠØ± Ù…ØªÙˆÙØ±ÙŠÙ†")

    # 5) Payment Method Donut
    if "Payment Method" in dff.columns and has_amount:
        pay_agg = dff.groupby("Payment Method", as_index=False)["Purchase Amount (USD)"].sum() \
                     .sort_values("Purchase Amount (USD)", ascending=False)
        fig_payment = px.pie(
            pay_agg, names="Payment Method", values="Purchase Amount (USD)",
            hole=0.5, color="Payment Method", color_discrete_sequence=PALETTE,
            title="Revenue Share by Payment Method"
        )
        fig_payment = style_pie(fig_payment)
    else:
        fig_payment = annotation_fig("Payment Method / Purchase Amount (USD) ØºÙŠØ± Ù…ØªÙˆÙØ±ÙŠÙ†")

    # 6) Treemap Category -> Item Purchased
    if "Category" in dff.columns and "Item Purchased" in dff.columns and has_amount:
        treemap_df = dff.groupby(["Category","Item Purchased"], as_index=False)["Purchase Amount (USD)"].sum()
        fig_treemap = px.treemap(
            treemap_df, path=["Category","Item Purchased"], values="Purchase Amount (USD)",
            color="Category", color_discrete_sequence=PALETTE, title="Treemap: Revenue by Category & Item"
        )
        fig_treemap.update_layout(margin=dict(l=20, r=20, t=60, b=20))
    else:
        fig_treemap = annotation_fig("Category / Item Purchased / Purchase Amount (USD) ØºÙŠØ± Ù…ØªÙˆÙØ±ÙŠÙ†")

    # 7) Box plot: Purchase Amount by Category
    if "Category" in dff.columns and has_amount:
        fig_box = px.box(
            dff, x="Category", y="Purchase Amount (USD)",
            color="Category", color_discrete_sequence=PALETTE,
            title="Order Value Distribution by Category (Box Plot)"
        )
        fig_box = style_common(fig_box, x_title="Category", y_title="Purchase Amount (USD)")
    else:
        fig_box = annotation_fig("Category / Purchase Amount (USD) ØºÙŠØ± Ù…ØªÙˆÙØ±ÙŠÙ†")

    # 8) Heatmap: Season x Category (Revenue)
    if "Season" in dff.columns and "Category" in dff.columns and has_amount:
        pivot = dff.pivot_table(index="Season", columns="Category",
                                values="Purchase Amount (USD)", aggfunc="sum", fill_value=0)
        fig_heat = px.imshow(
            pivot, color_continuous_scale="Blues", aspect="auto",
            title="Revenue Heatmap: Season Ã— Category"
        )
        fig_heat.update_layout(template="plotly_white", margin=dict(l=60, r=20, t=60, b=60))
    else:
        fig_heat = annotation_fig("Season Ã— Category (Revenue) ØºÙŠØ± Ù…ØªÙˆÙØ±")

    # 9) Scatter: Age vs Purchase Amount (marginals)
    if "Age" in dff.columns and has_amount:
        fig_scatter = px.scatter(
            dff, x="Age", y="Purchase Amount (USD)",
            color="Category" if "Category" in dff.columns else None,
            color_discrete_sequence=PALETTE,
            marginal_y="box", marginal_x="histogram",
            title="Age vs Order Value"
        )
        fig_scatter = style_common(fig_scatter, x_title="Age", y_title="Purchase Amount (USD)")
    else:
        fig_scatter = annotation_fig("Age / Purchase Amount (USD) ØºÙŠØ± Ù…ØªÙˆÙØ±ÙŠÙ†")

    # 10) Violin: Purchase Amount by Gender
    if "Gender" in dff.columns and has_amount:
        fig_violin = px.violin(
            dff, x="Gender", y="Purchase Amount (USD)",
            color="Gender", box=True, points="outliers",
            color_discrete_sequence=PALETTE, title="Order Value by Gender (Violin)"
        )
        fig_violin = style_common(fig_violin, x_title="Gender", y_title="Purchase Amount (USD)")
    else:
        fig_violin = annotation_fig("Gender / Purchase Amount (USD) ØºÙŠØ± Ù…ØªÙˆÙØ±ÙŠÙ†")

    return (
        kpi_cards,
        fig_cat, fig_gender, fig_rating, fig_season,
        fig_payment, fig_treemap, fig_box, fig_heat, fig_scatter, fig_violin,
        cat_vals, gen_vals, sea_vals
    )

# ================== Run ==================

   
# ==============================
if __name__ == '__main__':
    port = 8060
    print(f"-------------\n Dashboard Link: http://127.0.0.1:{port}/ \n-------------")
    app.run(debug=True, port=port) 

```
